**Data Merging and missing values treatment**

```python
#Import Necessary libraries
import pandas as pd
import numpy as np
```

**Merging the data**

```python
merged_df1 = pd.merge(appearances_df, games_df, on='game_id', how='inner')
```

```python
merged_df2 = pd.merge(merged_df1, games_events_df, on='game_id', how='inner')
```

```python
merged_df3 = pd.merge(merged_df2,players_df,left_on='player_id_x',right_on='player_id',how='inner')
```

```python
merged_df4 = pd.merge(merged_df3,games_lineups_df,on="player_id",how="inner")
```

```python
#check the missing values
merged_df4.isnull().sum()
```

```python
merged_df4.shape
```

```python
merged_df4.head()
```

```python
#missing value treatment
#Drop columns with more than 1000 missing values
columns_to_drop = merged_df4.columns[merged_df4.isnull().sum() > 1000]
merged_treated_df = merged_df4.drop(columns=columns_to_drop)

#Unneccessary columns
columns_to_drop =["competition_id_y","game_id_y",'player_id_y',"date_x","date_y","player_name_x"]
merged_treated_df=merged_treated_df.drop(columns_to_drop,axis=1)

#Treat remaining missing values
for column in merged_treated_df.columns:
    if merged_treated_df[column].isnull().sum() > 0:
        if  merged_treated_df[column].dtype == 'object':  # Categorical
           merged_treated_df[column].fillna(merged_treated_df[column].mode()[0], inplace=True)  # fill with mode
        else:  # Numerical
            merged_treated_df[column].fillna(merged_treated_df[column].mean(), inplace=True) #fill with mean
```

```python
#after missing values treatment check the missing values
merged_treated_df.isnull().sum()
```

```python
#Final data Shape
merged_treated_df.shape
```

```python
#check duplicate values
duplicates = merged_treated_df.duplicated()
print(duplicates)

# If you want to see only the duplicate rows
duplicate_rows = merged_treated_df[merged_treated_df.duplicated()]
print(duplicate_rows)
```

```python
#covert the file into CSV
merged_treated_df.to_csv('Football_data_merged_cleaned_data.csv', index=False)
```
